```python
import docx2txt
```

```python
import nltk
nltk.download()
```

```python
from nltk.tokenize import word_tokenize
from nltk.text import Text
```

```python
text_file=docx2txt.process("TOKYO_text.docx")
```

```python
text_file
```

```python
token=word_tokenize(text_file)
```

```python
token[:20]
```

```python
len(token)
```

```python
fdist1=nltk.FreqDist(token)
```

```python
word_freq= dict((word,freq) for word, freq in fdist1.items() if not word.isdigit())
```

```python
word_freq
```

```python
clean_words=[word for word in token if word.isalpha()]
```

```python
clean_words[:10]
```

```python
lower_words=[word.lower() for word in clean_words]
```

```python
lower_words
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
```

```python
print(stop_words)
```

```python
words_without_stop=[word for word in lower_words if not word in stop_words]
```

```python
words_without_stop
```

```python
len(words_without_stop)
```

```python
fdist2=nltk.FreqDist(words_without_stop)
```

```python
fdist2.most_common(20)
```

```python
textlist=Text(words_without_stop)
```

```python
type(textlist)
```

```python
textlist.concordance("tokyo")
```

```python
textlist.dispersion_plot(["tokyo"])
```
